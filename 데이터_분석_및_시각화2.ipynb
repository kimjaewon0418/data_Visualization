```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
#연령별 승선인원 카운트
plt.figure(figsize=(15,4))
sns.countplot(x='age', data=titanic)
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

```python
#1남자/여지/어린이 승선인원 시각화
#남자/여자/어린이 별 생존여부
sns.countplot(x='who', data=titanic,hue='alive')
```

```python
sns.rugplot(x='age', data=titanic,hue='alive')
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.displot(data=iris,x='petal_length',bins=20, rug=True,hue='species',kde=True)
```

```python
plt.figure(figsize=(12,8))
plt.subplot(221)
sns.boxplot(data=iris)
plt.subplot(222)
sns.violinplot(data=iris)
plt.subplot(223)
sns.swarmplot(data=iris,s=2)
plt.subplot(224)
sns.stripplot(data=iris)
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.subplot(221)
sns.boxplot(data=iris,x='species',y='petal_length')

plt.subplot(222)
sns.violinplot(data=iris,x='species',y='petal_length')

plt.subplot(223)
sns.swarmplot(data=iris,s=2,x='species',y='petal_length')

plt.subplot(224)
sns.stripplot(data=iris,x='species',y='petal_length')
plt.show()
```
