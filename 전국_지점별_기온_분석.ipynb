```python
#데이터 확인
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_1911 = pd.read_csv('/content/1911-1920 일별 기온.csv',encoding='cp949')
df_1921 = pd.read_csv('/content/1921-1930 일별 기온.csv',encoding='cp949')
df_1931 = pd.read_csv('/content/1931-1940 일별 기온.csv',encoding='cp949')
df_1941 = pd.read_csv('/content/1941-1950 일별 기온.csv',encoding='cp949')
df_1951 = pd.read_csv('/content/1951-1960 일별 기온.csv',encoding='cp949')
df_1961 = pd.read_csv('/content/1961-1970 일별 기온.csv',encoding='cp949')
df_1971 = pd.read_csv('/content/1971-1980 일별 기온.csv',encoding='cp949')
df_1981 = pd.read_csv('/content/1981-1990 일별 기온.csv',encoding='cp949')
df_1991 = pd.read_csv('/content/1991-2000 일별 기온.csv',encoding='cp949')
df_2001 = pd.read_csv('/content/2001-2010 일별 기온.csv',encoding='cp949')
df_2011 = pd.read_csv('/content/2011-2020 일별 기온.csv',encoding='cp949')
```

#2020년 지점별 기온 데이터 추출 및 전처리

```python
df_2020 = df[df['일시'].dt.year==2020].copy()
df_2020
```

```python
#지점별 데이터 개수 카운트
df_2020['지점명'].value_counts()
```

```python
#컬럼별 결측치 개수 확인
df_2020.isnull().sum()
```

```python
#결측치 데이터 확인
df_2020[(df_2020['평균기온(°C)'].isnull()) | (df_2020['최고기온(°C)'].isnull())]
```

```python
#결측치가 있는 데이터의 지점별 카운팅
df_2020[(df_2020['평균기온(°C)'].isnull()) | (df_2020['최고기온(°C)'].isnull())].value_counts('지점명')
```

```python
#결측치 제거
df_2020 = df_2020.dropna()
```

```python
df_2020.info()
```

```python
df_2020['지점명'].value_counts()
```

###지점별 연평균 기온 시각화

```python
df_2020_maen = df_2020.groupby('지점명')['평균기온(°C)'].mean().sort_values(ascending = False)
df_2020_maen
```

```python
import warnings
warnings.filterwarnings(action='ignore')
```

```python
#지점별 연 평균기온
plt.figure(figsize=(20,5))
plt.bar(df_2020_maen.index,df_2020_maen.values)
plt.xticks(rotation=90)

plt.axhline(df_2020_maen.max(),color='r',linestyle='--',label = 'max'+str(round(df_2020_maen.max(),1)))
plt.axhline(df_2020_maen.min(),color='b',linestyle='--',label = 'min'+str(round(df_2020_maen.min(),1)))
plt.axhline(df_2020_maen.mean(),color='g',linestyle='--',label = 'avg'+str(round(df_2020_maen.mean(),1)))

plt.legend()
plt.title('2020 market avg')
#서울 표시
plt.bar('서울',df_2020_maen.loc['서울'])

plt.show()
```

### 지점별 연 최저기온, 최고기온 시각화




```python
df_2020_cold = df_2020.groupby('지점명')['최저기온(°C)'].min().sort_values(ascending = False)
df_2020_cold
```

```python
#지점별 연 최저기온
plt.figure(figsize=(20,5))
plt.bar(df_2020_cold.index,df_2020_cold.values)
plt.xticks(rotation=90)
plt.title('2020 market cold')

plt.axhline(df_2020_cold.max(),color='r',linestyle='--',label = 'max'+str(round(df_2020_cold.max(),1)))
plt.axhline(df_2020_cold.min(),color='g',linestyle='--',label = 'min'+str(round(df_2020_cold.min(),1)))
plt.axhline(df_2020_cold.mean(),color='b',linestyle='--',label = 'mean'+str(round(df_2020_cold.mean(),1)))
plt.legend(loc=(0,1.01),ncol = 3,fontsize=12, edgecolor='k')


#서울 표시
plt.bar('서울',df_2020_cold.loc['서울'],color='r')
plt.text('서울', df_2020_cold.loc['서울']-1.5,'Seoul('+str(df_2020_cold.loc['서울'])+')',ha = 'center',fontsize=12)
plt.show()
```

```python
#지점별 연 최고기온
df_2020_hot = df_2020.groupby('지점명')['최고기온(°C)'].max().sort_values(ascending = False)
df_2020_hot
```

```python
#지점별 연 최저기온
plt.figure(figsize=(20,5))
plt.bar(df_2020_hot.index,df_2020_hot.values,color='pink')
plt.xticks(rotation=90)
plt.title('2020 market hot')

plt.axhline(df_2020_hot.max(),color='r',linestyle='--',label = 'max'+str(round(df_2020_hot.max(),1)))
plt.axhline(df_2020_hot.min(),color='g',linestyle='--',label = 'min'+str(round(df_2020_hot.min(),1)))
plt.axhline(df_2020_hot.mean(),color='b',linestyle='--',label = 'mean'+str(round(df_2020_hot.mean(),1)))
plt.legend(loc=(0,1.01),ncol = 3,fontsize=12, edgecolor='k')


#서울 표시
plt.bar('서울',df_2020_hot.loc['서울'],color='r')
plt.text('서울', df_2020_hot.loc['서울']+1,'Seoul('+str(df_2020_hot.loc['서울'])+')',ha = 'center',fontsize=12)
plt.show()
```
