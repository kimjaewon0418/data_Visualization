```python
df_1911 = pd.read_csv('/content/1911-1920 일별 기온.csv',encoding='cp949')
df_1921 = pd.read_csv('/content/1921-1930 일별 기온.csv',encoding='cp949')
df_1931 = pd.read_csv('/content/1931-1940 일별 기온.csv',encoding='cp949')
df_1941 = pd.read_csv('/content/1941-1950 일별 기온.csv',encoding='cp949')
df_1951 = pd.read_csv('/content/1951-1960 일별 기온.csv',encoding='cp949')
df_1961 = pd.read_csv('/content/1961-1970 일별 기온.csv',encoding='cp949')
df_1971 = pd.read_csv('/content/1971-1980 일별 기온.csv',encoding='cp949')
df_1981 = pd.read_csv('/content/1981-1990 일별 기온.csv',encoding='cp949')
df_1991 = pd.read_csv('/content/1991-2000 일별 기온.csv',encoding='cp949')
df_2001 = pd.read_csv('/content/2001-2010 일별 기온.csv',encoding='cp949')
df_2011 = pd.read_csv('/content/2011-2020 일별 기온.csv',encoding='cp949')
```

```python
df_seoul = df[df['지점명']== '서울'].copy()
```

```python
#결측치 확인
df_seoul.isnull().sum()
```

```python
df_seoul[(df_seoul['최저기온(°C)'].isnull()) | (df_seoul['최고기온(°C)'].isnull())]
```

```python
#결측치 제거
df_seoul = df_seoul.dropna()
```

```python
#결측치 제거 확인
df_seoul.isnull().sum()
```

```python
#연도별 데이터 개수 확인
df_seoul['일시'].dt.year.value_counts()
```

```python
#시각화로 데이터 확인
plt.plot(df_seoul['일시'],df_seoul['평균기온(°C)'])
plt.plot(df_seoul['일시'],df_seoul['최고기온(°C)'])
plt.plot(df_seoul['일시'],df_seoul['최저기온(°C)'])
plt.show()
```

```python
#데이터 확인
df_seoul[df_seoul['일시'].dt.year == 1951]
```

```python
#연간 평균기온의 평균 확인
df_seoul_mean = df_seoul.groupby(df_seoul['일시'].dt.year)['평균기온(°C)'].mean()
df_seoul_mean
```

```python
#데이터 시각화하여 데이터 확인
plt.figure(figsize=(15,5))
plt.plot(df_seoul_mean,'.')
plt.xticks(range(1911,2021,2),rotation=90)
plt.grid(ls=':')
plt.show()
```

```python
#이상치 제거
df_seoul_mean.drop(index = [1950,1953],inplace=True)
```

```python
plt.plot(df_seoul_mean,'g.-')
plt.title('Seoul average ')
plt.xticks(range(1911,2021,2),rotation=90)
plt.grid(ls=':')
plt.show()
```

```python
#서울시 연 최저기온 변화 확인
df_seoul_min = df_seoul.groupby(df_seoul['일시'].dt.year)['최저기온(°C)'].min()
df_seoul_min
```

```python
plt.plot(df_seoul_min,'b.-')
plt.title('Seoul min')
plt.show()
```

```python
#서울시 연 최고기온 변화 확인
df_seoul_max = df_seoul.groupby(df_seoul['일시'].dt.year)['최고기온(°C)'].max()
df_seoul_max
```

```python
df_seoul_max.drop(index=[1950,1953],inplace=True)
```

```python
plt.plot(df_seoul_max,'r.-')
plt.title('Seoul max')
plt.show()
```

```python
#서울시 연간 평균,최고,최저기온 한눈에 보기
plt.figure(figsize=(15,5))
plt.subplot(131)
plt.plot(df_seoul_mean,'g.-')
plt.title('Seoul average')
plt.xlabel('Year')
plt.ylabel('avg')
plt.grid(ls=':')

plt.subplot(132)
plt.plot(df_seoul_min,'b.-')
plt.title('Seoul Min')
plt.xlabel('Year')
plt.ylabel('min')
plt.grid(ls=':')

plt.subplot(133)
plt.plot(df_seoul_max,'r.-')
plt.title('Seoul Max')
plt.xlabel('Year')
plt.ylabel('max')
plt.grid(ls=':')

plt.show()
```

```python
#기온 변동 폭 확인
import warnings
warnings.filterwarnings(action='ignore')

plt.figure(figsize=(15,7))
plt.plot(df_seoul_mean,'g.-',label='avg')
plt.plot(df_seoul_min,'b.-',label='min')
plt.plot(df_seoul_max,'r.-',label= 'max')

plt.xticks(range(1910,2021,5),rotation=90)
plt.yticks(range(-25,42,5))
plt.grid(ls=':')
plt.legend()

plt.axhline(df_seoul_mean.min(),color='gray',ls='--')
plt.axhline(df_seoul_mean.max(),color='gray',ls='--')

plt.axhline(df_seoul_min.min(),color='gray',ls='--')
plt.axhline(df_seoul_min.max(),color='gray',ls='--')

plt.axhline(df_seoul_max.min(),color='gray',ls='--')
plt.axhline(df_seoul_max.max(),color='gray',ls='--')
plt.show()
```
